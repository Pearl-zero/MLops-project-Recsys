```python
import ast

# JSON 형식으로 저장된 장르 정보를 변환
def parse_genres(genres_str):
    if isinstance(genres_str, str):
        try:
            genres = ast.literal_eval(genres_str)
            return [genre['name'] for genre in genres] if isinstance(genres, list) else []
        except ValueError:
            return []
    elif isinstance(genres_str, list):
        return genres_str
    return []

movies_cleaned.loc[:, 'genres'] = movies_cleaned['genres'].fillna('[]').apply(parse_genres)

# One-hot encoding 사용 -> vectorize
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
genre_encoded = mlb.fit_transform(movies_cleaned['genres'])

# vectorize(genre) -> dataframe
genre_df = pd.DataFrame(genre_encoded, columns=mlb.classes_)
movies_with_genres = pd.concat([movies_cleaned, genre_df], axis=1)
```

```python
import os
import requests
from pyspark.sql import SparkSession
from pyspark.ml.recommendation import ALS  # ALS 임포트
import pandas as pd

# Python 경로 설정
os.environ['PYSPARK_PYTHON'] = 'C:/Users/admin/AppData/Local/Programs/Python/Python311/python.exe'
os.environ['PYSPARK_DRIVER_PYTHON'] = 'C:/Users/admin/AppData/Local/Programs/Python/Python311/python.exe'

# TMDB API 키 설정
TMDB_API_KEY = 'a33ff32c6de8efabceaafc6fdc0f107a'

# Spark session 생성
spark = SparkSession.builder.appName("CollaborativeFiltering").getOrCreate()

# 모델 학습 후 추천 결과를 DataFrame으로 변환
recommendations_df = als_model.recommendForAllUsers(10).toPandas()

# main_data의 id를 숫자형으로 변환
main_data['id'] = pd.to_numeric(main_data['id'], errors='coerce')

# recommendations_df의 movie_id를 숫자형으로 변환
recommendations_df['recommendations'] = recommendations_df['recommendations'].apply(
    lambda recs: [(int(movie_id), rating) for movie_id, rating in recs]
)

# NaN 값을 제거한 후 main_data의 id와 title을 dict type
main_data = main_data.dropna(subset=['id'])
id_to_title = dict(zip(main_data['id'], main_data['original_title']))

# links_small.csv
links = pd.read_csv('data/links_small.csv', encoding='utf8', low_memory=False)

# links_small.csv의 movieId와 tmdbId를 dict type
id_to_tmdbId = dict(zip(links['movieId'], links['tmdbId']))

# TMDB API 요청 결과를 캐싱할 딕셔너리
tmdb_cache = {}

# TMDB API로 영화 제목을 가져오는 함수 (캐시 적용)
def fetch_movie_title_from_tmdb(tmdb_id):
    if pd.notna(tmdb_id):
        # 이미 캐시에 있는 경우
        if tmdb_id in tmdb_cache:
            return tmdb_cache[tmdb_id]
        
        # 캐시에 없는 경우 API 호출
        url = f"https://api.themoviedb.org/3/movie/{int(tmdb_id)}?api_key={TMDB_API_KEY}&language=en-US"
        response = requests.get(url)
        if response.status_code == 200:
            data = response.json()
            title = data.get('title', f"Unknown Title (tmdbId: {tmdb_id})")
            tmdb_cache[tmdb_id] = title  # 캐시에 저장
            return title
        else:
            return f"Unknown Title (tmdbId: {tmdb_id})"
    else:
        return "Unknown Title"

# 추천 영화 ID와 영화 제목을 매핑하는 함수
def get_movie_titles(recommendations):
    titles = []
    for movie_id, _ in recommendations:
        if movie_id in id_to_title:
            # main_data에서 영화 제목을 찾은 경우
            titles.append(id_to_title[movie_id])
        elif movie_id in id_to_tmdbId:
            # links_small.csv에서 tmdbId로 영화 제목을 찾은 경우
            tmdb_id = id_to_tmdbId[movie_id]
            titles.append(fetch_movie_title_from_tmdb(tmdb_id))
        else:
            # 영화 제목을 찾지 못한 경우
            titles.append(f"Unknown Title (movie_id: {movie_id})")
    return titles

# 추천 영화 제목을 추가
recommendations_df['movie_titles'] = recommendations_df['recommendations'].apply(get_movie_titles)

# userId와 추천된 영화 제목만 출력
print(recommendations_df[['userId', 'movie_titles']])
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.sql import functions as F

# 예측
predictions = als_model.transform(ratings_spark)

# 평점을 기준으로 좋아하는 영화(positive)와 그렇지 않은 영화(negative)를 구분
# 3.5 이상의 평점을 positive로 설정
threshold = 3.5
predictions = predictions.withColumn("true_positive", F.when(predictions['rating'] >= threshold, 1).otherwise(0))
predictions = predictions.withColumn("pred_positive", F.when(predictions['prediction'] >= threshold, 1).otherwise(0))

# double로 변환
predictions = predictions.withColumn("true_positive", predictions["true_positive"].cast("double"))
predictions = predictions.withColumn("pred_positive", predictions["pred_positive"].cast("double"))

# Precision, Recall, F1 계산
evaluator_precision = MulticlassClassificationEvaluator(
    labelCol="true_positive", predictionCol="pred_positive", metricName="precisionByLabel")
evaluator_recall = MulticlassClassificationEvaluator(
    labelCol="true_positive", predictionCol="pred_positive", metricName="recallByLabel")
evaluator_f1 = MulticlassClassificationEvaluator(
    labelCol="true_positive", predictionCol="pred_positive", metricName="f1")

# 결과 계산
precision = evaluator_precision.evaluate(predictions)
recall = evaluator_recall.evaluate(predictions)
f1 = evaluator_f1.evaluate(predictions)

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")
```
